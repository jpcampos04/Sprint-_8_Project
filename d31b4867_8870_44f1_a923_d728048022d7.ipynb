## Prueba de hipótesis (Python)

 - Hipótesis nula:

**"La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare es la misma los sábados lluviosos".**

 - Hipótesis alternativa:

**"La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos".**

```python
# Carga el DataFrame
df_airport_trips = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
print('La información del DataFrame es:')
print()
df_airport_trips.info()
print()
print('Las columnas del DataFrame son:')
print(df_airport_trips.columns)
print()
print('El tamaño del DataFrame es:')
print(df_airport_trips.shape)
print()
print('Los primeros datos del DataFrame son:')
print(df_airport_trips.head())
print()
print('El DataFrame tiene los siguientes valores ausentes:')
print(df_airport_trips.isna().sum())
print()
print('El DataFrame tiene los siguientes duplicados:')
print(df_airport_trips.duplicated().sum())
print()
print(df_airport_trips[df_airport_trips.duplicated()])
```

El DataFrame **airport_trips** contiene los horarios, las condiciones meteorologicas y la duración del viaje. Se observa que posee tres columnas de las cuales los tipos de datos son incorrectos unicamente en la columna que contiene la fecha, el formato en que se nombran las columnas es correcto, no posee valores ausentes y los valores duplicados se deben a que las horas se redondean al igual que la duración del viaje por lo que se mantendran estos valores.

```python
# Cambia el tipo de dato a fecha
df_airport_trips['start_ts'] = pd.to_datetime(df_airport_trips['start_ts'], format='%Y-%m-%d %H:%M:%S')
#Extrae el día en una nueva columna
df_airport_trips['dt_day'] = df_airport_trips['start_ts'].dt.day
print(df_airport_trips.head())
print()
df_airport_trips.info()
```

```python
# Filtra la muestra por día sábado
df_airport_trips_weather = df_airport_trips[(df_airport_trips['dt_day'] == 4) | (df_airport_trips['dt_day'] == 11) | (df_airport_trips['dt_day'] == 18) | (df_airport_trips['dt_day'] == 25)]
print(df_airport_trips_weather)
```

<div class="alert alert-block alert-success">
<b>Celdas [13–14]</b> <a class="tocSkip"></a><br>
Transformas adecuadamente las fechas y seleccionas los sábados de forma precisa. Esta selección de muestras para filtrar por día específico es sólida y está bien ejecutada. Buen enfoque para preparar los datos antes de aplicar análisis estadístico.
</div>


```python
# Filtra la muestra por clima
good_weather_trips = df_airport_trips_weather[df_airport_trips_weather['weather_conditions'] == 'Good']['duration_seconds']
good_weather_trips_mean = df_airport_trips_weather[df_airport_trips_weather['weather_conditions'] == 'Good'].groupby('weather_conditions')['duration_seconds'].mean()
#print(good_weather_trips.head())
print(good_weather_trips_mean)

bad_weather_trips = df_airport_trips_weather[df_airport_trips_weather['weather_conditions'] == 'Bad']['duration_seconds']
bad_weather_trips_mean = df_airport_trips_weather[df_airport_trips_weather['weather_conditions'] == 'Bad'].groupby('weather_conditions')['duration_seconds'].mean()
#print(bad_weather_trips.head())
print(bad_weather_trips_mean)
```

```python
# Calcula la varianza de las muestras
good_weather_variance = np.var(good_weather_trips, ddof=1)
bad_weather_variance = np.var(bad_weather_trips, ddof=1)
print(good_weather_variance, 'varianza good.')
print(bad_weather_variance, 'varianza bad.')

# Prueba levene
levene_test = st.levene(good_weather_trips, bad_weather_trips)
print('valor p: ', levene_test.pvalue) # Si valor p > alpha 0.5 las varianzas son iguales por lo tanto True
# Si valor p < alpha 0.5 las varianzas son diferentes por lo tanto False

alpha = 0.05 # el nivel de significancia estadística crítica
# si el valor p es menor que alpha, rechazamos la hipótesis

good_vs_bad = st.ttest_ind(good_weather_trips, bad_weather_trips, equal_var=True) # realizar una prueba

print('valor p de la prueba t: ', good_vs_bad.pvalue) # extraer el valor p

if good_vs_bad.pvalue < alpha: # comparar el valor p con el umbral
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

<div class="alert alert-block alert-warning">
<b>Celdas [15–16]</b> <a class="tocSkip"></a><br>
La prueba de hipótesis está correctamente aplicada, pero sería recomendable eliminar valores con duración igual a 0 antes de ejecutar las pruebas estadísticas. Estos podrían sesgar la media y afectar la validez del test.
</div>


<div class="alert alert-block alert-success">
<b>Celda [16]</b> <a class="tocSkip"></a><br>
Excelente uso de Levene para validar la homogeneidad de varianzas antes de aplicar el t-test. Esto demuestra criterio técnico sólido en la elección de pruebas estadísticas. También interpretas correctamente el valor p y tomas decisiones fundadas en los resultados.
</div>


Las hipotesis nulas se formulan bajo la premisa que el promedio de las muestras son iguales(duración del viaje). Mientras que las hipotesis alternativas formula que existen diferencias entre el promedio de las muestras(duración del viaje).

El criterio que se utilizó para probar las hipotesis fueron que el nivel de varianza no puede exceder un 5% (un rango estadístico general) superior o inferior una muestra de otra.
